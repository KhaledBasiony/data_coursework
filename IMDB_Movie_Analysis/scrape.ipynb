```python
from bs4 import BeautifulSoup
import requests
import re
import pickle
```

```python
file = './List of currencies of the world by countries and territories.html'
with open(file, 'r') as f:
    soup = BeautifulSoup(f.read())
print(soup.prettify())
```

```python
next_val = {1:2, 2:3, 3:1}
ls = []
counter = 1
for td in soup.find_all('td'):
    txt = td.get_text()
    print(txt)
    if(len(txt) == 1):
        continue
    if(counter == 1):
        name = re.search('\(.+\)', txt)
        if name is None:
            name = txt
        else:
            name = txt[:name.start()-1]
    if(counter == 3):
        val = txt
        ls.append((name, val))
    counter = next_val[counter]
```

```python
ls
```

```python
country_currency = dict(ls)
```

```python
with open('country_currency', 'wb') as f:
    pickle.dump(country_currency, f)
```

```python
link = 'https://www.iban.com/exchange-rates'
r = requests.get(link)
soup = BeautifulSoup(r.content)
```

```python
print(soup.prettify())
```

```python
curr = []
conv = []
for i, td in enumerate(soup.find_all('td')):
    txt = td.get_text()
    if i%2 == 0:
        if i%4==0:
            curr.append(txt.strip())
        else:
            conv.append(txt.strip())
```

```python
currency_convert = dict(zip(curr, conv))
```

```python
len(currency_convert)
```

```python
with open('currency_convert', 'wb') as f:
    pickle.dump(currency_convert, f)
```
